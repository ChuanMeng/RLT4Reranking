 Efficiency-Effectiveness Tradeoff (EET)

```python

file_names=[
            "dl-19-passage.label-original-bm25-1000.rankllama-1000-ndcg@10.json",
            "dl-20-passage.label-original-bm25-1000.rankllama-1000-ndcg@10.json",
            #
            "dl-19-passage.label-original-splade-pp-ed-pytorch-1000.rankllama-1000-ndcg@10.json",
            "dl-20-passage.label-original-splade-pp-ed-pytorch-1000.rankllama-1000-ndcg@10.json",
            #
            "dl-19-passage.label-original-repllama-1000.rankllama-1000-ndcg@10.json",
            "dl-20-passage.label-original-repllama-1000.rankllama-1000-ndcg@10.json",
            #
            "dl-19-passage.label-original-bm25-1000.monot5-1000-ndcg@10.json",
            "dl-20-passage.label-original-bm25-1000.monot5-1000-ndcg@10.json",
            ]

def cal_eet(retrieval_result, reranking_result, k, alpha=None, beta=None):

  effectivenss = reranking_result - retrieval_result # [0, 1]
  if effectivenss<0:
      effectivenss=0.
  efficiency = math.exp(alpha*k)
  numerator = (1+beta**2)*(efficiency*effectivenss)
  denominator = (beta**2)*effectivenss+efficiency
  return numerator/denominator


for file_name in file_names:
  data_path = input_path+"/"+file_name
  setup = f"fig.eet-cutoff."+file_name.replace(".json",".pdf")

  if "ndcg@10" in file_name:
    metric = "nDCG@10"
    lower_bound = 0.4
  elif  "mrr@10" in file_name:
    metric = "MRR@10"
    lower_bound = 0.4
  elif  "map@100" in file_name:
    metric = "MAP@100"
    lower_bound = 0.2

  else:
    raise Exception

  if "dev" in file_name:
    lower_bound = 0.0

  with open(data_path, 'r') as r:
    q2result = json.load(r)


  xy={"k":[],"eet":[],"class":[]}

  for alpha in [-0.001]:
    for beta in [0,1,2]:

      qs_eet =[]

      for qid in q2result.keys():
        q_eet=[]
        retrieval_result = q2result[qid][0]
        for idx, reranking_result in enumerate(q2result[qid]):
          k=idx+1
          q_eet.append(cal_eet(retrieval_result, reranking_result, k, alpha=alpha, beta=beta))
        qs_eet.append(q_eet)

      eet_k =np.mean(np.array(qs_eet), axis=0)

      for idx, eet in enumerate(eet_k):
        xy["k"].append(idx+1)
        xy["eet"].append(eet)
        xy["class"].append(r"$\beta$={}, $\alpha$={}".format(beta,alpha))

  # aggregate
  df_xy = pd.DataFrame(xy)

  fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(4, 3))
  sns.set_style(style)

  sns.lineplot(ax=ax, data=df_xy, x="k", y="eet", hue='class')

  plt.legend()

  plt.xlim(0, 1000)
  plt.ylim(0, 0.5) # need to change
  #plt.title(name,fontsize=13)
  plt.ylabel("EET",fontsize=13)
  plt.xlabel("Re-ranking cut-off",fontsize=13)

  plt.tight_layout()
  #plt.show()
  plt.savefig(f"{output_path}/{setup}", format="pdf", bbox_inches="tight")
```
